```
import pandas as pd
import numpy as np
import sklearn
from sklearn.decomposition import TruncatedSVD
import warnings
```

```
books=pd.read_csv('Data/books.csv',sep=',',error_bad_lines=False)
books = books.iloc[:, :16]
books = books.drop(columns=['title', 'best_book_id', 'work_id', 'books_count', 'isbn', 'isbn13', 'original_publication_year','language_code','work_ratings_count','work_text_reviews_count'])
books.head(5)
```

```
ratings=pd.read_csv('Data/ratings.csv',sep=',',error_bad_lines=False)
ratings.head()
```

```
df=pd.merge(books,ratings,on='book_id')
df= df.drop_duplicates(['user_id','original_title'])
df= df.drop_duplicates(['user_id','book_id'])
df.head(10) #went down from 79701 to 79531 
df.shape #(79531, 8)
```

```
books_matrix = df.pivot_table(index = 'user_id', columns = 'original_title', values = 'rating').fillna(0)
books_matrix.shape #(28554, 794)
books_matrix.head()
```

```
X = books_matrix.values.T
X.shape
#Fitting the Model
SVD = TruncatedSVD(n_components=12, random_state=0)
matrix = SVD.fit_transform(X)
matrix.shape #(812, 12)
```

```
corr = np.corrcoef(matrix)
corr.shape
```

```
title = books_matrix.columns
title_list = list(title)
list(title_list)
```

```
idx = title_list.index("The Hitchhiker's Guide to the Galaxy")
corr_idx  = corr[idx]
list(title[(corr_idx >= 0.9)])
```
